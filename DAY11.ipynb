# CORRELATION 

```python
import pandas as pd
import matplotlib.pyplot as mp
from scipy.stats import pearsonr
d1 = pd.read_csv("general_data.csv")
d1=d1.dropna()
d1['Attrition'] = pd.factorize(d1.Attrition)[0]
```

# <span style='color:Red'> Correlation between Attrition and DistanceFromHome </span> 


```python
stats,p=pearsonr(d1.Attrition,d1.DistanceFromHome)
print("r:",stats)
print("p-value:",p)
print(mp.scatter(d1.Attrition,d1.DistanceFromHome))
```

# Conclusion: 
As r value is negative so Negative Correlation between Attrition and DistanceFromHome
